```python
import numpy as np
import matplotlib.pyplot as plt

data = np.linspace(1,10,20)
newFigure = plt.figure(figsize=(5,8))
newAxis = newFigure.add_axes([0.1,0.1,0.9,0.9])
newAxis.plot(data,data*2,label="Data * 2")
newAxis.plot(data,data*3,label="Data * 3")
newAxis.plot(data,data*4,label="Data * 4")
newAxis.plot(data,data*5,label="Data * 5")
newAxis.legend(loc=6)
newFigure.savefig("newfigure.png",dpi=100)
```

```python
data2 = np.linspace(10,100,20)
newFigure = plt.figure()
newAxis = newFigure.add_axes([0.1,0.1,0.9,0.9])
newAxis.plot(data2,data2*2,label="Data * 2")
newAxis.plot(data2,data2*3,label="Data * 3")
newAxis.plot(data2,data2*4,label="Data * 4")
newAxis.plot(data2,data2*5,label="Data * 5")
newAxis.legend(loc=2)
newFigure.savefig("newfigure1.png",dpi=150)
```
