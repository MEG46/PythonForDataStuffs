```python
import numpy as np
import matplotlib.pyplot as plt
arr1 = [14,20,25,52,54]
arr2 = [64,115,74,98,62]
age = np.array(arr1)
weight = np.array(arr2)
table = plt.plot(age,weight,"r")
plt.title("Age-Weight Table \"Red\"")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.show()
```
