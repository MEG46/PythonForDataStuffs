```python
import numpy as np
import matplotlib.pyplot as plt
data = np.linspace(1,10,20)
data1 = data ** 2
figure = plt.figure()
figureaxes1 = figure.add_axes([0.1,0.1,0.9,0.9])
figureaxes2 = figure.add_axes([0.2,0.5,0.3,0.3])
figureaxes1.plot(data,data1,"r--")
figureaxes1.set_xlabel("Numbers")
figureaxes1.set_ylabel("2nd Exponentials Of Numbers")
figureaxes1.set_title("Table in Table Example")

figureaxes2.plot(data1,data,"b")
figureaxes2.set_ylabel("Numbers")
figureaxes2.set_xlabel("2nd Exponentials Of Numbers")
figureaxes2.set_title("Table in Table Example")
plt.show()
```
