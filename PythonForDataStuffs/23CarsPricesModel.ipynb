```python
newdataFrame = newdataFrame[newdataFrame.year != 1970]
newdataFrame.groupby("year").mean()["price"]
```

```python
newdataFrame = newdataFrame.drop("transmission",axis=1)
newdataFrame.describe()
```

```python
import tensorflow as tf
y = newdataFrame["price"].values
x = newdataFrame.drop("price",axis=1).values
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=10)
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error,mean_squared_error
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential()
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))

model.compile(optimizer = "adam",loss = "mse")
model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=250,epochs=300)
```

```python
lostdata=pd.DataFrame(model.history.history)
lostdata.head()
```

```python
lostdata.plot()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
testGuess = model.predict(x_test)
mean_absolute_error(y_test,testGuess)
```

```python
plt.scatter(y_test,testGuess)
plt.plot(y_test,y_test,"r-*")
```
