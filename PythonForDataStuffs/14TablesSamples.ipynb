```python
import numpy as np
import matplotlib.pyplot as plt

data = np.linspace(0,10,20)
data1 = data ** 3
plt.title("Table 1")
plt.xlabel("Numbers")
plt.ylabel("3rd Exponentials of Numbers")
table = plt.plot(data,data1,"b--")
plt.show()
```
