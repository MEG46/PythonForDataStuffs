```python
import numpy as np
import matplotlib.pyplot as plt
data = np.linspace(1,10,20)
data1 = data ** 2
figure = plt.figure()
figureaxes = figure.add_axes([0.2,0.2,0.9,0.9])
figureaxes.plot(data,data1,"r--")
figureaxes.set_xlabel("Numbers")
figureaxes.set_ylabel("2nd Exponentials Of Numbers")
figureaxes.set_title("Figure To Table Example")
plt.show()
```
