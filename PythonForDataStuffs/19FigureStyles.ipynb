```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
data = np.linspace(1,10,20)
data1 = data ** 2
(myFigure, myAxis) = plt.subplots()
myAxis.plot(data,data1,color="darkcyan",alpha=0.7)
myAxis.plot(data1,data,color="orange",alpha=0.6)
```

```python
(myFigure, myAxis) = plt.subplots()
myAxis.plot(data,data+4,color="darkcyan",linewidth=1.0,linestyle="--")
myAxis.plot(data,data+6,color="orange",linewidth=5.0,linestyle=":")
myAxis.plot(data,data+8,color="darkcyan",linewidth=4.0,linestyle="-.")
myAxis.plot(data,data+14,color="red",linewidth=4.0,linestyle="--",marker="o",markersize=4,markerfacecolor="yellow")
myAxis.plot(data,data+12,color="green",linewidth=4.0,linestyle="--",marker="X",markersize=8,markerfacecolor="blue")
myAxis.plot(data,data+10,color="orange",linewidth=6.0)
myFigure.savefig("newfig2.png",dpi=200)
```
